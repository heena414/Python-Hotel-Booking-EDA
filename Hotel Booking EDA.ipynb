### Count Plot

#### A countplot is a type of bar plot that displays the count of observations in each categorical bin. It visualizes the frequency distribution of a categorical variable, with bars representing the count of occurrences for each category. The plot can also be segmented by additional categorical variables using color coding to show different groups or statuses.

### Grouped Bar Chart

### Does Coun Plot also known as Grouped Bar Chart?

#### Yes, a countplot can be considered a type of grouped bar chart. In a countplot, each category is represented by a bar, and when you add a `hue` parameter, it creates groups within each category, similar to how a grouped bar chart functions. The bars are grouped based on the levels of the `hue` variable, allowing for a visual comparison of subcategories within the main categories. This makes it an effective tool for visualizing counts of categorical data with an additional layer of grouping.

```python
# Extracting the month from 'reservation_status_date' and adding it as a new column 'month'
df['month']=df['reservation_status_date'].dt.month

# Creating a count plot to display the number of reservations per month, differentiated by cancellation status
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
ax1.legend(title='Reservation Status', loc='upper right', labels=['Not Canceled', 'Canceled'])
plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

plt.show()
```

```python
def get_count_from_column(df,column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index,'count':df_grpd.values})
  return df_grpd
```

```python
# plot a piechart from a group data

def plot_pie_chart_from_column(df,column_label,t1,exp):
  df_grpd = get_count_from_column(df,column_label)
  fig, ax = plt.subplots(figsize=(10,8))
  ax.pie(df_grpd.loc[:,'count'],labels = df_grpd.loc[:,'index'],autopct = '%1.2f%%', startangle=90,shadow = True,labeldistance = 1, explode =exp)
  plt.title(t1, bbox={'pad':6})
  ax.axis('equal')
  plt.legend()
  plt.show()
```

```python
exp1 = [0.08,0.08]
plot_pie_chart_from_column(df,'hotel','Booking percentage of Hotel',exp1)
```
